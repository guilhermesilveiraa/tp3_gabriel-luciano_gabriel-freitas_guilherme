O PostgreSQL realiza suas transações respeitando as propriedas ACID.


Podemos definir esta organização da estrutura físcia nos itens a seguir:
*   Arquivo de Dados: O PostgreSQL armazena os dados em arquivos no sistema de  arquivos do sistema operacional.

#Tarefa 13: Crie um índice para o atributo ‘v’ e realize consultas e criação de índice

Qual o tempo gasto para realizar uma consulta para um valor (tendo a tabela 100000 tuplas)?
Qual o tempo gasto para recriar um índice para o atributo ‘v’?

Remova a tabela ‘t’ e crie novamente com 1.0000.000 de tuplas

Qual o tempo gasto para realizar uma consulta para um valor específico?
Qual o tempo gasto para recriar um índice para o atributo ‘v’?


````python
import psycopg2
import time

conn = psycopg2.connect(
    host="postgres",
    database="tp3_ggg",
    user="postgres",
    password="postgres"
)

# Etapa 1 da tarefa 13 - Criando índice
cria_indice = """
    DROP INDEX IF EXISTS indice;
    CREATE INDEX indice ON t(v);
"""

cur = conn.cursor()
cur.execute(cria_indice)
conn.commit()

# Etapa 2 da tarefa 13 - Tempo gasto para realizar uma consulta para um valor (tendo a tabela 100000 tuplas)
tempo_consulta = """
    EXPLAIN ANALYZE SELECT * FROM t WHERE v = 1000000;
"""
cur.execute(tempo_consulta)
conn.commit()

res = cur.fetchall()

for r in res:
    print(r)
    print('-----------------------------')

# Etapa 3 da tarefa 13 - Tempo gasto para recriar um índice para o atributo 'v'
inicia_contagem = time.time()
recriacao = """
    DROP INDEX indice;
    CREATE INDEX indice ON t(v);
"""
cur.execute(recriacao)
conn.commit()
termina_contagem = time.time()
tempo_recriacao = termina_contagem - inicia_contagem

print("Tempo de recriação do índice:", tempo_recriacao, "segundos")

#Etapa 4 da tarefa 13 - Remova a tabela ‘t’ e crie novamente com 1.0000.000 de tuplas
recria_tabela = """
    DROP TABLE IF EXISTS t;
    CREATE TABLE t (k serial PRIMARY KEY, v integer);
    INSERT INTO t(v)
    SELECT trunc(random() * 10) FROM generate_series(1, 1000000);
    SELECT * FROM t ORDER BY v ASC LIMIT 10
    """
cur.execute(recria_tabela)
conn.commit

res_recria = cur.fetchall()
print('Recriando tabela agora com 1000000 elementos:')

for r in res_recria:
    print(r)

#Etapa 5 da tarefa 13 - Qual o tempo gasto para realizar uma consulta para um valor específico?
tempo_consulta2 = """
    EXPLAIN ANALYZE SELECT * FROM t WHERE v = 1000000;
"""
cur.execute(tempo_consulta2)
conn.commit()

res_2 = cur.fetchall()
print('Tempo de consulta a um valor depois de recriar tabela:')
for r in res_2:
    print(r)
    print('-----------------------------')


cur.close()
conn.close()
````
